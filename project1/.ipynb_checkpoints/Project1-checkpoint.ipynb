```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
# Get html data from url
r = requests.get('https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares')
```

```python
root = BeautifulSoup(r.content,'html5lib')
```

```python
# Get table data
table = root.find('table')
```

```python
# Create new Table
new_table = pd.DataFrame(columns=['rank', 'x_class', 'date', 'region', 'start_time', 'max_time', 'end_time', 'movie'], index = range(0,50))

# Put data in table
row_marker = 0
for row in table.find_all('tr'):
    column_marker = 0
    columns = row.find_all('td')
    row_marker += 1
    for column in columns:
        if (row_marker > 1):
            new_table.iat[row_marker-2,column_marker] = column.get_text()
            column_marker += 1
```

```python
new_table
```

```python
# Drop the last column
new_table = new_table.drop('movie', 1)
```

```python
new_table
```

```python
new_table.dtypes
```

```python
# Combine the date column with the time columns to convert to datetime format
df_st = pd.to_datetime(new_table['date'] + ' ' + new_table['start_time'])
df_mt = pd.to_datetime(new_table['date'] + ' ' + new_table['max_time'])
df_et = pd.to_datetime(new_table['date'] + ' ' + new_table['end_time'])

# Add the new columns of the datetimes
new_table['start_datetime'] = df_st
new_table['max_datetime'] = df_mt
new_table['end_datetime'] = df_et

# Drop the unneeded columns
new_table = new_table.drop('date', 1)
new_table = new_table.drop('start_time', 1)
new_table = new_table.drop('max_time', 1)
new_table = new_table.drop('end_time', 1)
```

```python
# Change the order of the columns
new_table = new_table[['rank', 'x_class', 'start_datetime', 'max_datetime', 'end_datetime', 'region']]

# Replace all instances of '-' with '<NA>' as stated by the prompt
new_table = new_table.replace('-', '<NA>')
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
r = requests.get('http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html')
```

```python
root = BeautifulSoup(r.content,'html5lib')
```

```python
table = root.find('pre')
```

```python
row_marker = 0
content = table.get_text()
lines = content.split('\n')
for i in range(0,11):
    lines.pop(0);
lines.pop(len(lines)-1)
lines.pop(len(lines)-1)
lines.pop(len(lines)-1)
new_table = pd.DataFrame(columns=['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency', 'flare_location', 'flare_region','flare_classification', 'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed'], index = range(0,len(lines)))
for line in lines:
    cols = line.split(' ')
    while '' in cols:
        cols.remove('')
    col_marker = 0
    while col_marker < 14:
        new_table.iat[row_marker, col_marker] = cols[col_marker]
        col_marker += 1
    row_marker += 1
```

```python
new_table
```

```python
#helper method

def isInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False
    
#START OF STEP 4

# Replace all instances of '----' with 'NaN' as stated by the prompt
new_table = new_table.replace('----', 'NaN').replace('-----', 'NaN')

#Populate Halo_flare column
new_table['is_halo'] = new_table['cme_angle'].map(lambda x: x == 'Halo')

#Replace "Halo" with 'Na'
new_table = new_table.replace('Halo', 'NA')

#populate widthlowerbound column
new_table['width_lower_bound'] = new_table['cme_width'].map(lambda x: str(x)[0] == '>')

#remove any non-numeric part of the width column.
new_table['cme_width'] = new_table['cme_width'].map(lambda x: x if isInt(x) else "")
```

```python
# Combine the date column with the time columns to convert to datetime format
new_table['start_datetime'] = pd.to_datetime(new_table['start_date'] + ' ' + new_table['start_time'])
new_table['end_datetime'] = pd.to_time(new_table['end_time'])
```

```python
# Drop the unneeded columns
new_table = new_table.drop('start_date', 1)
new_table = new_table.drop('start_time', 1)
new_table = new_table.drop('end_date', 1)
new_table = new_table.drop('end_time', 1)

new_table = new_table[['start_datetime', 'end_datetime', 'start_frequency', 
                       'end_frequency', 'flare_location', 'flare_region','flare_classification', 
                       'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed', 'is_halo', 
                       'width_lower_bound' ]]

new_table
```
